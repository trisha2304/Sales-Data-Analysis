# Sales Data Analysis Using Python

## Project Overview
This project performs basic sales data analysis using Python, Pandas, and Matplotlib. The objective is to analyze sales records, identify product and monthly sales trends, and visualize the results using charts.

## Technologies Used
- Python
- Pandas
- Matplotlib
- Jupyter Notebook

```python
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries imported successfully!")
```

```python
# Load the sales dataset
df = pd.read_csv("sales_data.csv")

# Display the first 5 rows
df.head()
```

```python
# Check basic information about the dataset
df.info()
```

```python
# Check for missing values
print("Missing values:")
print(df.isnull().sum())
```

```python
# Calculate total sales for each order
df["Total_Sales"] = df["Quantity"] * df["Price"]

# Display updated dataset
df.head()
```

```python
# Calculate total revenue
total_revenue = df["Total_Sales"].sum()

print("Total Revenue: ₹", total_revenue)
```

```python
# Calculate total sales for each product
product_sales = df.groupby("Product")["Total_Sales"].sum()

print("Sales by Product:")
print(product_sales)
```

```python
# Find the product with the highest sales
highest_product = product_sales.idxmax()
highest_sales = product_sales.max()

print("Product with Highest Sales:", highest_product)
print("Highest Sales Amount: ₹", highest_sales)
```

```python
# Visualize sales by product
product_sales.plot(kind="bar")

plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales (₹)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate total sales for each month
monthly_sales = df.groupby("Month")["Total_Sales"].sum()

print("Monthly Sales:")
print(monthly_sales)
```

```python
# Arrange months in chronological order
month_order = [
    "January", "February", "March", "April",
    "May", "June", "July", "August"
]

monthly_sales = monthly_sales.reindex(month_order)

print(monthly_sales)
```

```python
# Find the month with the highest sales
best_month = monthly_sales.idxmax()
best_month_sales = monthly_sales.max()

print("Highest Sales Month:", best_month)
print("Sales Amount: ₹", best_month_sales)
```

```python
# Visualize monthly sales
monthly_sales.plot(kind="line", marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales (₹)")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
